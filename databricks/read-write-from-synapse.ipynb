### Read table from synapse data using lake gen2

synapse & data laek gen2 within a custom vnet

```
%fs ls /mnt/bifconnect/
```

```
blobStorage = "gdwpocdls.dfs.core.windows.net"
blobContainer = "bifconnect"
blobAccessKey = dbutils.secrets.get(scope = 'BIFC-asagdwpocws-etl', key = 'StorageAccount-gdwpocdls-AccessKey')

spark.conf.set(f"fs.azure.account.key.{blobStorage}",blobAccessKey)

tempDir = "abfss://" + blobContainer + "@" + blobStorage +"/tempDirs"

_url = dbutils.secrets.get(scope = 'BIFC-asagdwpocws-etl', key = 'Synapse-asagdwpocetl-ConnectionString');


# Load data from an Azure Synapse query.
df1 = (spark.read.format("com.databricks.spark.sqldw")
       .option("query", "select * from ACC_MART.Dim_Suite")
       .option( "useAzureMSI","True")
       .option("url", _url)
       .option("tempdir", tempDir)
       .load())

display(df1)
```

### Write table to synapse

```
df = (spark
    .read
    .format('csv')
    .option('header', True)
    .load('/mnt/blob/cpc/dbo.D21_3S_TRANS_OIL_AZURE.csv'))

display(df)
```

```
(df.write
    .format("com.databricks.spark.sqldw")
    .option("useAzureMSI", "true")
    .option("dbTable", "D21_3S_TRANS_OIL_AZURE")
    .option("url", _url)
    .option("tempdir", tempDir)
    .save()
)
```

```
df1 = (spark.read.format("com.databricks.spark.sqldw")
       .option("query", "select * from D21_3S_TRANS_OIL_AZURE")
       .option( "useAzureMSI","True")
       .option("url", _url)
       .option("tempdir", tempDir)
       .load())
 
display(df1)
```
